# **Extract Region Data From MCA Files**

This notebook shows how to extract region data from MCA files.

```python
%load_ext autoreload
%autoreload 2
```

```python
import sys
sys.path.append('../')

import os

from src.mca.region_from_mca import get_region
from src.utils.block_dictionary import get_block_id_dictionary, get_block_color_dictionary

os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

DATA_PATH = '../data/'
REGION_FILE_PATH = f'{DATA_PATH}region.mca'

CHUNK_X, CHUNK_Z = 15, 25
SECTION_Y = 9
CLUSTER_SIZE = 3
```

```python
block_id_dict = get_block_id_dictionary()
block_color_dict = get_block_color_dictionary()
```

```python
region = get_region(REGION_FILE_PATH, block_id_dict=block_id_dict, parallelize_chunks=True)
region.display(block_id_dict, block_color_dict)
```

```python
chunk = region.get_chunk(CHUNK_X, CHUNK_Z)
chunk.display()
```

```python
section = chunk.get_section(SECTION_Y)
section.display()
```

```python
section = chunk.get_section(SECTION_Y + 1)
section.display()
```

```python
cluster = region.get_cluster(CHUNK_X- CLUSTER_SIZE // 2, SECTION_Y- CLUSTER_SIZE // 2, CHUNK_Z- CLUSTER_SIZE // 2, CLUSTER_SIZE)
cluster.display()
```
